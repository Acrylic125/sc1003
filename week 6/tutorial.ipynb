````python

sample_string = "Hello, World! Welcome to Python programming."

upper_string = sample_string.upper()
print(f"Uppercase: {upper_string}") # HELLO, WORLD! WELCOME TO PYTHON PROGRAMMING.

lower_string = sample_string.lower()
print(f"Lowercase: {lower_string}") # hello, world! welcome to python programming.

capitalized_string = sample_string.capitalize() 
print(f"Capitalized: {capitalized_string}") # Hello, world! welcome to python programming.

position = sample_string.find("World")
print(f"Position of 'World': {position}") # Position of 'World': 7

replaced_string = sample_string.replace("World", "Universe")
print(f"Replaced string: {replaced_string}") # Hello, Universe! Welcome to Python programming.

words_list = sample_string.split()
print(f"List of words: {words_list}") # List of words: ['Hello,', 'World!', 'Welcome', 'to', 'Python', 'programming.']

joined_string = " ".join(words_list)
print(f"Joined string: {joined_string}") # Joined string: Hello, World! Welcome to Python programming.
````

````python
temperatures = [72, 68, 75, 70, 69, 74, 73]
print(f"Recorded temperatures: {temperatures}") # Recorded tempratures [72, 68, 75, 70, 69, 74, 73]

temperatures.sort() # [68, 69, 70, 72, 73, 74, 75]
print(temperatures) # [68, 69, 70, 72, 73, 74, 75]
new_temperature = 71
temperatures.append(new_temperature) # [68, 69, 70, 72, 73, 74, 75, 71]
print(f"After appending new temperature {new_temperature}: {temperatures}") # After appending new temperature 71: [68, 69, 70, 72, 73, 74, 75, 71]

max_temperature = max(temperatures) # 75
print(f"The maximum temperature recorded is: {max_temperature}") # The maximum temperature recorded is: 75
min_temperature = min(temperatures)
print(f"The minimum temperature recorded is: {min_temperature}") # The minimum temperature recorded is: 68

pre_temperatures = [51, 53, 56]
pre_temperatures.append(temperatures) # [51, 53, 56, [68, 69, 70, 72, 73, 74, 75, 71]]
print(f"after apend pre_temperatures:{pre_temperatures}") # after apend pre_temepratures:[51, 53, 56, [68, 69, 70, 72, 73, 74, 75, 71]]

aft_temperatures = [95, 97, 94]
for temp in aft_temperatures:
    temperatures.append(temp) 
# temperatures = [68, 69, 70, 72, 73, 74, 75, 71, 95, 97, 94]
print(f"After appending temperatures for another week: {temperatures}") # After appending temperatures for another week: [68, 69, 70, 72, 73, 74, 75, 71, 95, 97, 94]
average_temperature = sum(temperatures)/len(temperatures) # 78.0
print("The average temperature is:", average_temperature) # The average temperature is: 78.0
````

# Q4
## Problem
Write a Python program, in the fewest number of lines possible, which creates a list of all the square numbers: x^2 (where 1<=x<=100) that are divisible by 3.

## Solution
```python
x2 = [x ** 2 for x in range(1, 101) if x ** 2 % 3 == 0]
```

````python
# Define a dictionary representing a collection of books in a library
library = {
    "978-0143127741": {
        "title": "To Kill a Mockingbird",
        "author": "Harper Lee",
        "year": 1960,
        "genre": "Fiction"
    },
    "978-0439023481": {
        "title": "The Hunger Games",
        "author": "Suzanne Collins",
        "year": 2008,
        "genre": "Dystopian"
    },
    "978-0307277671": {
        "title": "The Road",
        "author": "Cormac McCarthy",
        "year": 2006,
        "genre": "Post-apocalyptic"
    }
}

# Print the entire library dictionary
print("Library collection:")
for isbn, book in library.items():
    # ISBN: 978-0143127741, Book: { 'title': 'To Kill a Mockingbird', 'author': 'Harper Lee',  'year': 1960, 'genre': 'Fiction'}
    # ISBN: 978-0439023481, Book: { 'title': 'The Hunger Games', 'author': 'Suzanne Collins', 'year': 2008, 'genre': 'Dystopian' }
    # ISBN: 978-0307277671, Book: { 'title': 'The Road', 'author': 'Cormac McCarthy', 'year': 2006, 'genre': 'Post-apocalyptic' }
    print(f"ISBN: {isbn}, Book: {book}")

# Access information about a specific book by its ISBN
isbn_to_lookup = "978-0143127741"
book_info = library.get(isbn_to_lookup, "Book not found")
# \nDetails of the book with ISBN 978-0143127741: { 'title': 'To Kill a Mockingbird', 'author': 'Harper Lee',  'year': 1960, 'genre': 'Fiction'} 
print(f"\nDetails of the book with ISBN {isbn_to_lookup}: {book_info}")

# Add a new book to the library
new_book_isbn = "978-0553573404"
new_book = {
    "title": "A Game of Thrones",
    "author": "George R. R. Martin",
    "year": 1996,
    "genre": "Fantasy"
}
library[new_book_isbn] = new_book
# \nAdded new book with ISBN 978-0553573404: { 'title': 'A Game of Thrones', 'author': 'George R. R. Martin', 'year': 1996, 'genre': 'Fantasy' }
print(f"\nAdded new book with ISBN {new_book_isbn}: {library[new_book_isbn]}")

# Remove a book from the library
isbn_to_remove = "978-0307277671"
removed_book = library.pop(isbn_to_remove, "Book not found")
# \nRemoved book with ISBN 978-0307277671: { 'title': 'The Road', 'author': 'Cormac McCarthy', 'year': 2006, 'genre': 'Post-apocalyptic' }
print(f"\nRemoved book with ISBN {isbn_to_remove}: {removed_book}")

# Update the details of an existing book
isbn_to_update = "978-0439023481"
library[isbn_to_update]["year"] = 2009  # Updating the year of publication
# \nUpdated book details with ISBN 978-0439023481: { 'title': 'The Hunger Games', 'author': 'Suzanne Collins', 'year': 2009, 'genre': 'Dystopian' }
print(f"\nUpdated book details with ISBN {isbn_to_update}: {library[isbn_to_update]}")

# Search for books by a specific author
author_to_search = "Harper Lee"
books_by_author = [book for book in library.values() if book["author"] == author_to_search]
# \nBooks by Harper Lee: [{ 'title': 'To Kill a Mockingbird', 'author': 'Harper Lee',  'year': 1960, 'genre': 'Fiction'}]
print(f"\nBooks by {author_to_search}: {books_by_author}")

# Filter books by genre
genre_to_filter = "Fiction"
books_in_genre = {isbn: book for isbn, book in library.items() if book["genre"] == genre_to_filter}
# \nBooks in the genre 'Fiction': {'978-0143127741' { 'title': 'To Kill a Mockingbird', 'author': 'Harper Lee',  'year': 1960, 'genre': 'Fiction'}}
print(f"\nBooks in the genre '{genre_to_filter}': {books_in_genre}")
````
