# Q5
## Problem
Write a program that asks the user for the number of boys and that of girls in a class. The program should calculate and display the percentage of boys and girls in the class. A sample run is as follows:
Enter the number of boys: 65
Enter the number of girls: 77
Boys: 46%
Girls: 54%

Design the algorithm and use flowchart to present.
Write the Python program. (optional)

## Solution

````python
boys = int(input("Enter the number of boys: ")) # Throws a ValueError if the input is not an integer
girls = int(input("Enter the number of girls: "))

total = boys + girls
percent_boys = (boys * 100) / total
percent_girls = (girls * 100) / total
print(f"Boys: {percent_boys}%")
print(f"Girls: {percent_girls}%")
````
